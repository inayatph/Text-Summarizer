```python
from transformers import BartTokenizer, BartForConditionalGeneration

# pre-trained BART tokenizer and model from the Hugging Face Transformers library.
# BART is a transformer-based model architecture specifically designed for text generation tasks, including text summarization.
# bidirectional and auto regressive transformer is pretrained model made by facebook AI
```

```python
# Load pre-trained BART tokenizer and model
tokenizer = BartTokenizer.from_pretrained('facebook/bart-large-cnn')

# initialize tokenizer object from BART model named facebook/bart , tokenizer converts input into tensors

model = BartForConditionalGeneration.from_pretrained('facebook/bart-large-cnn')

# initialized model object of BART for class BARTforConditionalGeneration
```

```python
# max_length = 1024 to handle longer texts
# return_tensors='pt' is pytorch tensor
# truncation = True to truncate the input if it exceeds the maximum length supported by the model.

# Generating Summary: We use the generate method of the BART model to generate the summary. We pass the input tensors to the model
# along with parameters like num_beams (number of beams for beam search), min_length (minimum length of the generated summary),
#  max_length (maximum length of the generated summary),
# and early_stopping (whether to stop generation early if all beam hypotheses have reached the minimum length).

# summary id = as the output is tensor containing token ids , these are stored in summary ids. since there can be multiple summary cause
# num beams > 1 so summary ids[0] take 0th position summary
# skip special token = skipping spcl token such as pad , sep= seperator and cls whic is not needed in summarization
```

```python
def summarize_text(input_text):
    # Tokenize input text
    inputs = tokenizer([input_text], max_length=1024, return_tensors='pt', truncation=True)

    # Generate summary
    summary_ids = model.generate(inputs['input_ids'], num_beams=4, min_length=30, max_length=100, early_stopping=True)

    # Decode summary
    summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)
    return summary
```

```python
# This code provides a basic implementation of text summarization using the BART model. You can further customize the summarization
# process by adjusting parameters such as num_beams, min_length, and max_length based on your requirements.

# Additionally, you can fine-tune the pre-trained model on domain-specific data for better summarization performance.
```

```python
#beam=in a sequence based predictions the next token is selected using previous tokens  and the best candidates that can be used as
# the next token are called beams . num beams gives u the no. of beams used for next token . higher the count of num beams higher is
# the computation resulting in better prediction
```

```python

# Create a Gradio interface
demo = gr.Interface(fn=summarize_text, inputs="text",
                     outputs="text", title="Text Summarization",
                     description="Enter text to be summarized.")
demo.launch()
```
